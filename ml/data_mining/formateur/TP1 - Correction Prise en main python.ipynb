## Partie 1 : Rappels des bases de Python

```python
# ---

# ### 1.1 Variables et types de base
# Une variable est un espace mémoire qui contient une valeur.

# Exemple :
nom = "Alice"  # Chaîne de caractères
age = 25       # Entier
prix = 19.99   # Flottant
est_etudiant = True  # Booléen

# Affichons les valeurs des variables :
print("Nom :", nom)
print("Age :", age)
print("Prix :", prix)
print("Est étudiant :", est_etudiant)

# ---
```

```python
# ---

# ### 1.2 Listes et boucles
# Une liste permet de stocker plusieurs valeurs.

# Exemple :
notes = [15, 12, 18, 10, 14]  # Une liste de notes

# Afficher chaque note avec une boucle :
for note in notes:
    print("Note :", note)

# Calculons la moyenne des notes :
moyenne = sum(notes) / len(notes)
print("Moyenne des notes :", moyenne)

# ---
```

```python
# ---

# ### 1.3 Fonctions
# Une fonction permet de réutiliser un bloc de code.

# Exemple :
def calculer_moyenne(liste):
    """
    Cette fonction calcule la moyenne d'une liste de nombres.
    """
    # Return average
    return sum(liste) / len(liste)

# Utilisation de la fonction :
moyenne_notes = calculer_moyenne(notes)
print("Moyenne calculée avec la fonction :", moyenne_notes)

# ---
```

## Partie 2 : Manipulation de données avec Pandas

```python
# Pandas est une bibliothèque très utilisée pour manipuler et analyser les données tabulaires.

# ---

# ### 2.1 Importer les bibliothèques
import pandas as pd  # Import de pandas pour la manipulation des données
import matplotlib.pyplot as plt  # Import de matplotlib pour les visualisations

# Configuration de l'affichage
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 200)

# ---
```

```python
# ---

# ### 2.2 Lecture d'un fichier CSV
# On va lire un fichier CSV contenant des données fictives sur des ventes.

# **NB :** Assurez-vous d'avoir un fichier "ventes.csv" dans le même dossier que ce notebook.
# Exemple de contenu du fichier "ventes.csv" :
# Produit,Quantite,Prix
# Pomme,10,0.5
# Orange,15,0.8
# Banane,7,0.3

# Lecture du fichier CSV :
chemin_fichier = "fagor.xlsx"  # Remplacez par le chemin de votre fichier

# Créons un DataFrame Pandas :
df = pd.read_excel(chemin_fichier)

# Affichons les 5 premières lignes du DataFrame :
print("Aperçu des données :")
print(df.head())

# ---
```

```python
# ---

# ### 2.3 Exploration des données
# Pandas nous permet d'analyser rapidement les données.

# Afficher les informations générales sur le DataFrame :
print("\nInformations sur les données :")
print(df.info())

# Quelques statistiques descriptives :
print("\nStatistiques descriptives :")
print(df.describe())

# ---
```

```python
# ---

# ### 2.4 Manipulation des données
# On peut créer de nouvelles colonnes, filtrer des données, etc.

# Créer une nouvelle colonne 'Diff_Angelua1_Alpha' (Angelua 1 - Alpha) :
df['Diff_Angelua1_Alpha'] = df['Angelua 1']-df['Alpha']

# Afficher le DataFrame modifié :
print("\nDonnées avec la colonne 'Diff_Angelua1_Alpha' :")
print(df.head(10))

# ---
```

```python
# ---

# ### 2.5 Visualisation simple
# On utilise Matplotlib pour créer des graphiques simples.

# Graphique : Quantités vendues par produit
plt.figure(figsize=(8, 5))  # Taille de la figure
plt.bar(df['Angelua 1'], df['SPRINGBACK (DEG)'], color='skyblue')
plt.title('SPRINGBACK (DEG) selon \n l\'angle final Angelua 1 obtenu')
plt.xlabel('Angelua 1')
plt.ylabel('SPRINGBACK (DEG)')
plt.show()

# ---
```

```python
# ### Exercice 3 : Visualisation des totaux
# # Créez un graphique (barres) montrant le total des ventes par produit.

# # Réponse :

df = pd.read_csv('/content/cars.csv')

# Créer le groupby
total_ventes = df.groupby('Fuel_Type')['Car_Name'].count().reset_index()

# Créer un graphique en barres avec plt
plt.bar(total_ventes['Fuel_Type'], total_ventes['Car_Name'], color='lightgreen')
plt.title('Total des ventes par produit')
plt.xlabel('Produits')
plt.ylabel('Total des ventes')
plt.show()

# Créer un graphique en barres avec seaborn
import seaborn as sns
plt.figure(figsize=(8, 6))  # Taille de la figure
sns.barplot(x='Fuel_Type', y='Car_Name', data=total_ventes, palette='viridis', hue='Fuel_Type', legend=False)

# Ajout de titres et labels
plt.title("Total des ventes par type de carburant", fontsize=14)
plt.xlabel("Type de carburant", fontsize=12)
plt.ylabel("Total des ventes", fontsize=12)

# Afficher le graphique
plt.show()
```
